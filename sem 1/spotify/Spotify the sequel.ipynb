```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("spotify_songs.csv")
df.info()
```

```python
sns.histplot(data=df, x="track_popularity", bins=50)
```

```python
summary = {
    "Max": df["track_popularity"].max(),
    "Q3": df["track_popularity"].quantile(0.75),
    "Median": df["track_popularity"].median(),
    "Q1": df["track_popularity"].quantile(0.25),
    "Min":df["track_popularity"].min()
}
summary
```

```python
sns.boxplot(data=df, x="track_popularity")
```

```python
def avg(col):
    average = col.mean()
    return average
```

```python
avg(df["tempo"])
```

```python
def standard_deviation(dataframe, column):
    std = dataframe[column].std()
    return std
y = standard_deviation(df, "danceability")
```

```python
sns.histplot(data=df, x="danceability", bins=50)
```

```python
x = df["danceability"].mean()
```

```python
print("A majority of songs fall within: ", (x+y), "and", (x-y))
```

```python
df["playlist_genre"].unique()
```

```python
df_pop = df[df["playlist_genre"] == "pop"]
df_rap = df[df["playlist_genre"] == "rap"]
```

```python
sns.histplot(data=df_pop, x="speechiness", bins=50, color="green")
sns.histplot(data=df_rap, x="speechiness", bins=50, color="blue")
```

```python
pop_mean = df_pop["speechiness"].mean()
rap_mean = df_rap["speechiness"].mean()
print("Pop: ", pop_mean, "Rap:", rap_mean)
```
