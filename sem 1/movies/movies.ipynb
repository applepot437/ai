```python
def get_revenue_budget(df):
    '''
    Uses the movie id to get data from the "Details" endpoint
    Returns a dataframe of revenue and budget data
    '''
    data = {
        "id":[],
        "budget": [],
        "revenue": []
    }

    for i in df["id"]:
        data["id"].append(i)
        url = "https://api.themoviedb.org/3/movie/"+str(i)+"?language=en-US"

        headers = {
            "accept": "application/json",
            "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI4Nzk4M2ZiZTk3Njk4OGE5ZTIyMzMwZWJmMDMwYTFhZSIsIm5iZiI6MTcyNzk3MTk0OS45MTA1MjcsInN1YiI6IjYzNzQyNGMyNDZlNzVmMDA5MGE3YmRhYyIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.ZsTpMVaVlF8ioN3-WiYEatqWmnGQhCKlILQv4Xi6Eec"
        }

        response = requests.get(url, headers=headers)
        response = response.json()
        data["budget"].append(response["budget"])
        data["revenue"].append(response["revenue"])
        
    return pd.DataFrame(data)
```

```python
df = df[['budget', 'revenue']].dropna()
sns.lmplot(x='budget', y='revenue', data=df)
```

```python
df['budget'] = df['budget'].astype(float)
df['revenue'] = df['revenue'].astype(float)
```

```python
n = len(df)
sum_x = df['budget'].sum()
sum_y = df['revenue'].sum()
sum_x_squared = (df['budget'] ** 2).sum()
sum_xy = (df['budget'] * df['revenue']).sum()
```

```python
slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
intercept = (sum_y * sum_x_squared - sum_x * sum_xy) / (n * sum_x_squared - sum_x ** 2)
```

```python
budget = 5_000_000
predicted_revenue = slope * budget + intercept

predicted_revenue
```

```python
df['ROI'] = (df['revenue'] - df['budget']) / df['budget']
```

```python
n = len(df)
sum_x = df['budget'].sum()
sum_y = df['ROI'].sum()
sum_x_squared = (df['budget'] ** 2).sum()
sum_xy = (df['budget'] * df['ROI']).sum()
```

```python
slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
intercept = (sum_y * sum_x_squared - sum_x * sum_xy) / (n * sum_x_squared - sum_x ** 2)
```

```python
denominator = (n * sum_x_squared - sum_x ** 2)
if denominator == 0:
    print("Error: Zero denominator in slope/intercept calculation")
else:
    slope = (n * sum_xy - sum_x * sum_y) / denominator
    intercept = (sum_y * sum_x_squared - sum_x * sum_xy) / denominator
```
